```python
import numpy as np
import matplotlib.pyplot as plt
```

### EJERCICIO 1:

```python
def metodo_de_la_potencia(A,k):
    # PRE: A es matriz cuadrada y k un entero positivo
    v = np.random.rand(len(A)) #Tomo como vector inicial un vector cualquiera con dimensión n
    a = [] #Vector de dimensión k en el que almacenaré las aproximaciones de autovalores que calcula el programa
    for i in range(k):
        Av = A@v
        v = Av/np.linalg.norm(Av,2)
        r = (np.transpose(v)@A@v) / (np.transpose(v)@v)
        a.append(r)
    # POST: a_i converge al autovalor de A de mayor módulo
    return a
```

```python
A = np.array([[3,1,7],[-4,1,-9],[4,0,4]])

autovalores = np.linalg.eigvals(A)
autovectores = np.linalg.eig(A)[1]
print(autovalores)
print(autovectores)
print(A)

metodo_de_la_potencia(A,50)
```

```python
A = np.random.rand(10,10) #Matriz random de 10x10

autovalores = np.linalg.eigvals(A)
#autovectores = np.linalg.eig(A)[1]
print(autovalores)
#print(autovectores)

metodo_de_la_potencia(A,50)
```

### EJERCICIO 2:

### A)

```python
A = np.random.rand(100,100) #Matriz random de 100x100

autovalores = np.linalg.eigvals(A)
#autovectores = np.linalg.eig(A)[1]
print(autovalores)
#print(autovectores)

aproximaciones = metodo_de_la_potencia(A,100)
print(aproximaciones)
```

### B)

```python
pasos = np.linspace(1,100,100)
print(pasos)
```

```python
plt.figure(figsize=(15, 4))
plt.title("Velocidad de convergencia en función de los pasos")
plt.plot(pasos, aproximaciones, "o", color = "red")
plt.xlabel("PASOS")
plt.ylabel("APROXIMACIÓN");
plt.figure(figsize=(20, 4))

plt.show()
```

El método converge de forma bastante rápida, y en menos de 5 pasos usualmente llega a una aproximación del autovalor de mayor módulo bastante estable (o al menos así se vio en las pruebas ejecutadas)

### EJERCICIO 3:

### A)

```python
I = np.identity(100)

C = np.random.rand(100,100)
A = 0.5*(C+np.transpose(C)) #Matriz simétrica
B = A + 500*I

aproximaciones = metodo_de_la_potencia(B,100)
```

```python
plt.figure(figsize=(15, 4))
plt.title("Velocidad de convergencia en función de los pasos")
plt.plot(pasos, aproximaciones, "o", color = "red")
plt.xlabel("PASOS")
plt.ylabel("APROXIMACIÓN");
plt.figure(figsize=(20, 4))

plt.show()
```

### B)

```python
def calcular_errores_de_paso(aproximaciones, B):
    e = [] #Vector de errores
    lambda_max = np.linalg.eigvals(B)[0] #Autovalor de mayor módulo de B
    for i in range(len(B)):
        error_de_paso = abs(lambda_max - aproximaciones[i])
        e.append(error_de_paso)
    return e

errores = calcular_errores_de_paso(aproximaciones,B)
print(errores)
```

<font color='purple'>Comentario: la matriz puede tener autovalores negativos. En caso de que el autovalor de mayor módulo sea negativo, no estarías comparando el autovalor hallado con el de mayor módulo. </font>

```python
pasos = np.linspace(1,100,100)

plt.figure(figsize=(15, 4))
plt.title("Módulo de error de aproximación en función de los pasos")
plt.plot(pasos, errores, "*", color = "blue")
plt.xlabel("PASOS")
plt.ylabel("ERROR DE APROXIMACIÓN");
plt.figure(figsize=(20, 4))

plt.show()
```

Se observa que el error en función del número de iteraciones es una función logarítmica decreciente, que se acerca asintóticamente al eje X (es decir, la magnitud del error cometido es cada vez menor).

<font color='purple'>Comentario: Se trata más bien de una función exponencial o de una hipérbola.</font>

### C)

```python
log_errores = [np.log(error) for error in errores]
print(log_errores)
```

```python
np.sort(np.linalg.eigh(B)[0])
```

```python
lambda_1 = np.sort(np.linalg.eigh(B)[0])[len(np.linalg.eigh(B)[0]) - 1] #Tomo autovalor de módulo más grande
lambda_2 = np.sort(np.linalg.eigh(B)[0])[len(np.linalg.eigh(B)[0]) - 2] #Tomo autovalor de segundo módulo más grande
print(lambda_1)
print(lambda_2)

x = pasos
y = 2*np.log(lambda_2/lambda_1)*x + log_errores[0]

plt.figure(figsize=(15, 4))
plt.title("Logaritmo del módulo de error de aproximación en función de los pasos")
plt.plot(pasos, log_errores, "*", color = "blue")
plt.plot(x, y, color = "green")
plt.xlabel("PASOS")
plt.ylabel("LOG DE ERROR DE APROXIMACIÓN");
plt.figure(figsize=(20, 4))

plt.show()
```

<font color='purple'>Comentario: como antes, no estás tomando el módulo de los autovalores.</font>

Ahora, con los logaritmos de los errores de aproximación, podemos decir con seguridad que esta es una función lineal decreciente y no solo eso, sino que además el error decrece con cada paso del método. Además, vemos que la función propuesta (y(X)) se corresponde bastante bien (de hecho se superponen casi totalmente con los valores experimentales) con el comportamiento esperado para el módulo del error cometido con cada iteración.

<font color='purple'>Comentario: al graficar el logaritmo del error en función de las iteraciones, se obtiene una función lineal. Esto permite concluír que la función del error en función de las iteraciones es efectivamente una exponencial.</font>
