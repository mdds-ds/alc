# Mínimos cuadrados e interpolación polinomial

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Ejemplo:** Una pelota es arrojada hacia arriba. Una cámara de fotos toma fotos cada 0.1 segundos y a partir de esas fotos se calcula la altura de la pelota en cada instante.
En el archivo `altura_pelota.csv` se encuentran las mediciones obtenidas.

Sabiendo que la altura de la pelota depende del tiempo en forma cuadrática 
$$h(t) = c_0 + c_1 t + c_2 t^2,$$ estimar los valores de $c_0, c_1, c_2$ ajustando los datos por mínimos cuadrados.

Según la función que obtuvo, ¿en que instante golpeará la pelota el piso?

a) Usamos las ecuaciones normales.

```python
import pandas as pd   
# Los datos están en español! 
datos = pd.read_csv("altura_pelota.csv", sep=";", decimal=",")   # dataFrame
```

```python
# Convertimos los datos a np.array
datosNP = datos.to_numpy()
print(datosNP.shape)
```

```python
# Graficamos
plt.scatter(datosNP[:,0], datosNP[:,1])
```

a) Resolvemos el problema mediante las ecuaciones normales
$$
A^tA c = A^ty
$$
para una matriz $A$ apropiada.


```python
ti = datosNP[:,0]
yi = datosNP[:,1]

# Ajustamos una función f(t) = c0 + c1 t + c2 t^2
# Construimos la matriz A con columnas 1, t, t^2
unos = np.ones(len(ti))

A = np.c_[unos, ti, ti**2]

# Resolvemos el sistema
At = np.transpose(A)
c = np.linalg.solve(At @ A, At @ yi)
print(c)
```

```python
# Damos vuelta el vector de coeficientes para construir el polinomio poly1d
cUD = np.flipud(c)
print(cUD)

p = np.poly1d(cUD)
print(p)
```

```python
# Graficamos
plt.scatter(ti, yi, c = "g")
t = np.linspace(0,13, 100)
plt.plot(t, p(t))
```

```python
# Raíces
p.r
```

b) Lo resolvemos usando `np.polyfit`

```python
# polyfit(x, y, n) calcula el polinomio de grado n que mejor ajusta los datos (x_i, y_i) por mínimos cuadrados
# Nos devuelve el vector de coeficientes empezando por la potencia de grado más alto
d = np.polyfit(ti, yi, 2)
print(d)
```

```python
# Lo convertimos a polinomio y graficamos
p2 = np.poly1d(d)

plt.scatter(ti, yi, c = "g")
t = np.linspace(0,13, 100)
plt.plot(t, p2(t))
```

## Interpolación polinomial

Interpolamos la función 
$$
f(x) = x^{\frac23}
$$
para $n+1$ puntos equiespaciados en el intervalo $[-1,1]$.

```python
def f(x): 
    return(abs(x)**(2/3))
```

```python
# Graficamos
x_plot = np.linspace(-1,1,1001)
plt.plot(x_plot, f(x_plot))
```

```python
# Construimos el polinomio interpolador de grado n usando polyfit
n = 20
xi = np.linspace(-1, 1, n+1)
c = np.polyfit(xi, f(xi), n)
print(c)
```

```python
p = np.poly1d(c)
```

```python
plt.plot(x_plot, f(x_plot))
plt.plot(x_plot, p(x_plot))
```

```python
# Podemos calcular el error máximo numéricamente:
err = max(abs(f(x_plot) - p(x_plot)))
print(err)
```

```python
## Interpolación lineal a trozos
```

```python
# Graficamos
x_plot = np.linspace(-1,1,1001)
plt.plot(x_plot, f(x_plot))
```

```python
# Construimos el interpolador lineal a trozos 
n = 4
xi = np.linspace(-1, 1, n+1)
y_vals = np.interp(x_plot, xi, f(xi))
plt.plot(x_plot, y_vals)
plt.plot(x_plot, f(x_plot))
```

```python
# Podemos calcular el error máximo numéricamente:
err = max(abs(f(x_plot) - y_vals))
print(err)
```
